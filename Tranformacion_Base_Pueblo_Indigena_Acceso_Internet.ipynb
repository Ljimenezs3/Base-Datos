```python
url = "https://raw.githubusercontent.com/Omaira85/BASES-DE-DATOS-SEG-SEMESTRE/main/Pueblo%20indigena1.csv"
puebloindigena = pd.read_csv(url, sep=';', encoding = 'unicode_escape')
puebloindigena
```

## **<font color='#3498DB'>Limpieza</font>**

### **Eliminación de tildes**

```python
import unicodedata

def strip_accents(s):
  if (type(s) == str):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')
  return s




for column in puebloindigena.keys():
  puebloindigena[column] = puebloindigena[column].map(strip_accents)

puebloindigena
```

### **Cambio nombre de columnas**

```python
puebloindigena.columns = ['CODIGO', 'DEPARTAMENTO', 'POBLACION_TOTAL', 'PDP_MAYA','PDP_GARIFUNA','PDP_XINCA', 'PDP_AFRODESCENDIENTE/CREOLE/AFROMESTIZO', 'PDP_LADINA(O)','PDP_EXTRANJERA(O)' ]
```

### **Se remplazan espacios en blanco por ningún valor**

```python
def remover_espacio(cadena):
  return cadena.replace(" ", "", regex = True)
```

```python
puebloindigena.iloc[:, 2:9] = remover_espacio(puebloindigena.iloc[:, 2:9])
```

### **Se cambian tipos de datos de object a int64**

```python
puebloindigena = puebloindigena.astype({'POBLACION_TOTAL': 'int64', 'PDP_MAYA': 'int64', 'PDP_GARIFUNA': 'int64', 'PDP_XINCA': 'int64', 'PDP_AFRODESCENDIENTE/CREOLE/AFROMESTIZO': 'int64', 'PDP_LADINA(O)': 'int64', 'PDP_EXTRANJERA(O)': 'int64'})
```

## **<font color='#3498DB'>Cargue Dataframe Acceso Internet</font>**

>  Características generales de la población. Censo 2018: Este dataframe contiene la información de la población de 7 años o más por uso de celular, computadora y/o internet, de acuerdo al municipio

```python
accesointernet = pd.read_excel("/content/drive/MyDrive/Base de Datos/Acceso Internet.xlsx")
accesointernet
```

```python
import unicodedata

def strip_accents(s):
  if (type(s) == str):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')
  return s




for column in accesointernet.keys():
  accesointernet[column] = accesointernet[column].map(strip_accents)

accesointernet
```
